```python
import nltk
import re
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer

nltk.download("stopwords")
from nltk.corpus import stopwords

data = pd.read_csv('tripadvisor_hotel_reviews.csv')
data.head()

```

```python
data.info()
```

```python
data['Review'][0]
```

```python
data['review_lowercase'] = data['Review'].str.lower()

data.head( )

```

```python
en_stopwords = stopwords.words("english")
en_stopwords.remove('not')
```

```python
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
print(len(data["review_no_stopwords"][0]),len(data["review_lowercase"][0]))
data["review_no_stopwords"][0]
```

```python
data['review_no_stopword_no_punct'] = data.apply(lambda x: re.sub(r"[*]","star", x['review_no_stopwords']), axis=1)
data[['review_no_stopwords','review_no_stopword_no_punct']]
```

```python
data['review_no_stopword_no_punct'] = data.apply(lambda x: re.sub(r"[^\w\s]", "", x['review_no_stopword_no_punct']), axis=1)
data.head()
```

```python
data['tokenized'] =  data.apply(lambda x: word_tokenize(x['review_no_stopword_no_punct']), axis=1)
data['tokenized'][0]
```

```python
ps= PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens] )
```

```python
data[['tokenized','stemmed']]
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized'] =  data['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens] )
```

```python
data['lemmatized'][0]
```

```python
# We sum all the tokens fro all the review into a single list for review
tokens_clean = sum(data['lemmatized'], []) #<-- starting with empty list
```

```python
unigram = pd.Series(nltk.ngrams(tokens_clean,1)).value_counts()
print(unigram)
```

```python
bigram = pd.Series(nltk.ngrams(tokens_clean,2)).value_counts()
print(bigram)
```

```python
trigram = pd.Series(nltk.ngrams(tokens_clean,3)).value_counts()
print(trigram)
```
