```python
import nltk
import re
import spacy
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

bbc_data =  pd.read_csv("bbc_news.csv")
```

```python
titles = pd.DataFrame(bbc_data["title"])
```

```python
titles.head()
```

# Cleaning

```python
titles["lowercase"] = titles['title'].str.lower()
```

```python
en_stopword = stopwords.words('english')
titles['no_stopwords'] = titles['lowercase'].apply(
        lambda x:' '.join([word for word in x.split() if word not in (en_stopword)]))
# For each title in the lowercase column, split the title into individual words, remove any word that appears in the English stopword list (en_stopword), then join the remaining words back into a string separated by spaces.
```

```python
titles['no_stopwords_no_punct'] = titles.apply(lambda x: re.sub(r"([^\w\s])","", x['no_stopwords']), axis=1 )
```

```python
# #For each Row (AXIS 1) in the titles DataFrame, apply word_tokenize to the title column and store the list of tokens in a new column called tokens_raw.
# titles['tokens_raw'] = titles.apply(lambda x: word_tokenize(x['title']), axis= 1) 

# # For each Row (AXIS 1) with no stopwords nor punctuation, apply word_tokenize and store the tokens in a new column called tokens_clean.
# titles['tokens_clean']= titles.apply(lambda x: word_tokenize(x['no_stopwords_no_punct']), axis=1 ) 

# # Updated - easier to read and use:
titles['tokens_raw'] = titles['title'].apply(word_tokenize)

titles['tokens_clean'] = titles['no_stopwords_no_punct'].apply(word_tokenize)

```

```python
lemmatizer = WordNetLemmatizer()
titles["tokens_clean_lemmatized"] = titles['tokens_clean'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
titles.head()
```

```python
tokens_raw_list = sum(titles['tokens_raw'], []) #unpacking each list of each row into a single list
tokens_clean_list = sum(titles['tokens_clean_lemmatized'], [])
```

```python
nlp = spacy.load('en_core_web_sm')
```

## Raw

```python
spacy_doc = nlp(' '.join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns= ['token','pos_tag'])

for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{"token":token.text, "pos_tag":token.pos_}] )], ignore_index=True)
```

```python
pos_df_counts = pos_df.groupby(['token','pos_tag']).size().reset_index(name='counts').sort_values(by= 'counts', ascending=False)
pos_df_counts
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == 'NOUN']
nouns
```

```python
verbs = pos_df_counts[pos_df_counts.pos_tag == 'VERB']
verbs
```

```python
adj = pos_df_counts[pos_df_counts.pos_tag == 'ADJ']
adj
```

# NER

```python
ner_df =  pd.DataFrame(columns= ["token", 'ner_tag'])

for token in spacy_doc.ents:
    if pd.isna(token.label_) is False:
        ner_df = pd.concat([ner_df, pd.DataFrame.from_records([{'token': token.text, 'ner_tag': token.label_}] )], ignore_index= True)
```

```python
ner_df.head()
```

```python
ner_df_count = ner_df.groupby(['token','ner_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
ner_df_count
```
